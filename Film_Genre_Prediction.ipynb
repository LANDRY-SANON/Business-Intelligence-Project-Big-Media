```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import ast
```

```python
data_to_use['PE_Genre'] = data_to_use['PE_Genre'].apply(ast.literal_eval)
```

```python
unique_genres = set()
data_to_use['PE_Genre'].apply(lambda genres: unique_genres.update(genres))
print(f"Unique Genres ({len(unique_genres)}): {unique_genres}")
```

```python
cleaned_genres = set(genre.strip() for genre in unique_genres)

print(f"Cleaned Unique Genres ({len(cleaned_genres)}): {cleaned_genres}")
```

```python

genre_counts = {}

for genres in data_to_use['PE_Genre']:
    for genre in genres:
        if genre in genre_counts:
            genre_counts[genre] += 1
        else:
            genre_counts[genre] = 1

# Display the genre counts
print(genre_counts)
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

def predict_genre_with_gpt2(text, genres):
    genres_list = ', '.join(genres)
    prompt = f"Based on this synopsis, predict the right genre from the following list . There may be one or many correct genre : {genres_list}. Synopsis: {text}\nGenre:"
    inputs = tokenizer(prompt, return_tensors='pt', padding=True, truncation=True, max_length=512)
    print(inputs["input_ids"].shape[1])
    outputs = model.generate(**inputs, max_length=inputs["input_ids"].shape[1] + 10, do_sample=False)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    predicted_genre = response.split('Genre:')[-1].split('.')[0].strip()
    return predicted_genre



```

```python
synopsis = "Over the course of several years, two convicts form a friendship, seeking consolation and, eventually, redemption through basic compassion."

genre_prediction = predict_genre_with_gpt2(synopsis, cleaned_genres)
print(f"Predicted Genre: {genre_prediction}") #Drama
```

```python
synopsis = "Don Vito Corleone, head of a mafia family, decides to hand over his empire to his youngest son Michael. However, his decision unintentionally puts the lives of his loved ones in grave danger."
genre_prediction = predict_genre_with_gpt2(synopsis, cleaned_genres)
print(f"Predicted Genre: {genre_prediction}") #'Crime', 'Drama'
```

```python
def one_shot_genre_prediction(example_text, example_genre, target_text, genres):
    # Prepare the prompt with an example
    genres_list = ', '.join(genres)
    prompt = f"Example: Synopsis: {example_text} Genre: {example_genre}\nBased on this synopsis, predict the right genre from the following list . There may be one or many correct genre : {genres_list}. Synopsis: {target_text}\nGenre:"
    inputs = tokenizer(prompt, return_tensors='pt', padding=True, truncation=True, max_length=512)
    print(inputs["input_ids"].shape[1])
    outputs = model.generate(**inputs, max_length=inputs["input_ids"].shape[1] + 10,  do_sample=False)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    predicted_genre = response.split('Genre:')[-1].split('.')[0].strip()
    return predicted_genre

example_text = "An anime adaptation of the Hindu epic the Ramayana, where Lord Ram combats the wicked king Ravana.",
example_genre = "Animation, Action, Adventure"

```

```python
target_text = "Over the course of several years, two convicts form a friendship, seeking consolation and, eventually, redemption through basic compassion."

genre_prediction = one_shot_genre_prediction(example_text, example_genre, target_text, cleaned_genres)
print(f"Predicted Genre: {genre_prediction}") #Drama
```

```python

target_text = "Don Vito Corleone, head of a mafia family, decides to hand over his empire to his youngest son Michael. However, his decision unintentionally puts the lives of his loved ones in grave danger."
genre_prediction = one_shot_genre_prediction(example_text, example_genre, target_text, cleaned_genres)
print(f"Predicted Genre: {genre_prediction}")  #'Crime', 'Drama'
```

```python
def four_shot_genre_prediction(examples, target_text, genres):
    genres_list = ', '.join(genres)
    prompt = "Examples:\n"
    for text, genre in examples:
        prompt += f"Synopsis: {text} Genre: {genre}\n"
    prompt += f"Based on these synopses, predict the right genre from the following list: There may be one or many correct genre  {genres_list}. Synopsis: {target_text}\nGenre:"

    inputs = tokenizer(prompt, return_tensors='pt', padding=True, truncation=True, max_length=512)

    outputs = model.generate(**inputs,  max_length=inputs["input_ids"].shape[1] + 10 , do_sample=False)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    predicted_genre = response.split('Genre:')[-1].split('.')[0].strip()
    return predicted_genre

examples = [
    ("A young wizard finds himself competing in a hazardous tournament between rival schools of magic, but he is distracted by recurring nightmares.", "Fantasy"),
    ("A mafia family patriarch contemplates his legacy and succession plans amidst growing rivalries and an impending law enforcement crackdown.", "Crime"),
    ("In a post-apocalyptic world, a lone warrior seeks vengeance against a tyrannical leader who controls the last of the world's water supplies.", "Action"),
    ("Two unlikely friends journey across a magical forest to find a legendary creature said to grant wishes.", "Adventure")
]

```

```python
target_text = "Over the course of several years, two convicts form a friendship, seeking consolation and, eventually, redemption through basic compassion."
genre_prediction = four_shot_genre_prediction(examples, target_text, cleaned_genres)
print(f"Predicted Genre: {genre_prediction}")  #Drama
```

```python
target_text =  "Don Vito Corleone, head of a mafia family, decides to hand over his empire to his youngest son Michael. However, his decision unintentionally puts the lives of his loved ones in grave danger."
genre_prediction = four_shot_genre_prediction(examples, target_text, cleaned_genres)
print(f"Predicted Genre: {genre_prediction}")    #'Crime', 'Drama'
```

```python
import torch
from torch.utils.data import DataLoader, Dataset
from transformers import GPT2Tokenizer, GPT2ForSequenceClassification, GPT2Config
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer
from torch.nn import BCEWithLogitsLoss


tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

inputs = tokenizer(data_to_use['PE_Synopsis'].tolist(), truncation=True, padding=True, max_length=512)

mlb = MultiLabelBinarizer()
genre_labels = mlb.fit_transform(data_to_use['PE_Genre'])
num_labels = len(mlb.classes_)

config = GPT2Config.from_pretrained("gpt2", pad_token_id=tokenizer.eos_token_id, num_labels=num_labels)
model = GPT2ForSequenceClassification.from_pretrained("gpt2", config=config)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)


class GenreDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx], dtype=torch.float32)
        return item

    def __len__(self):
        return len(self.labels)

dataset = GenreDataset(inputs, genre_labels)
train_size = int(0.9 * len(dataset))
test_size = len(dataset) - train_size
train_dataset, test_dataset = torch.utils.data.random_split(dataset, [train_size, test_size])
train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=4, shuffle=False)

loss_function = BCEWithLogitsLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)
model.train()

for epoch in range(3):
    for batch in train_loader:
        inputs, labels = batch['input_ids'].to(device), batch['labels'].to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = loss_function(outputs.logits, labels)
        loss.backward()
        optimizer.step()
    print(f"Epoch {epoch} Loss: {loss.item()}")

model.eval()
total_eval_accuracy = 0
for batch in test_loader:
    inputs, labels = batch['input_ids'].to(device), batch['labels'].to(device)
    with torch.no_grad():
        outputs = model(inputs)
        logits = torch.sigmoid(outputs.logits)
        predictions = (logits > 0.5).int()
        total_eval_accuracy += (predictions == labels).sum().item()

accuracy = total_eval_accuracy / (len(test_dataset) * num_labels)
print(f"Accuracy: {accuracy:.4f}")
```
